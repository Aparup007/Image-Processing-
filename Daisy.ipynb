```python
import skimage 
from skimage import color,io,data
import matplotlib
from matplotlib import pyplot as plt
from skimage.feature import daisy
test=skimage.io.imread(r'C:\Users\user\AppData\Local\Programs\Python\Python37\Lib\site-packages\skimage\data\color.png')
test=color.rgb2gray(test)
plt.figure(figsize=(10,8))
plt.imshow(test,cmap='gray')
```

```python
descs_array,descs_image=daisy(test,step=200,radius=20,visualize=True)
descs_num=descs_array.shape[0]*descs_array.shape[1]
descs_num
plt.figure(figsize=(12,10))
plt.title('DAISY descriptors',fontsize=18)
plt.imshow(descs_image)
```
