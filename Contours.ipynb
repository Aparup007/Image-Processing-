```python
import skimage
from skimage import color,io,measure
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from skimage.util import view_as_blocks
test=skimage.io.imread(r'C:\Users\user\AppData\Local\Programs\Python\Python37\Lib\site-packages\skimage\data\horse.png')
plt.imshow(test)
```

```python
test=color.rgb2gray(test)
plt.imshow(test, cmap='gray')
```

```python
contours=measure.find_contours(test,0.01)
len(contours)
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,12),sharex=True,sharey=True)

ax=axes.ravel()
ax[0].imshow(test, cmap='gray')
ax[0].set_title('Original Image',fontsize=20)

ax[1].imshow(test,cmap='gray',interpolation='nearest')
ax[1].set_title('Contour',fontsize=20)

for n, contour in enumerate(contours):
    ax[1].plot(contour[:,1],contour[:,0],linewidth=5)

plt.show()
```
